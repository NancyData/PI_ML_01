```python
#SE HACE EL MERGE POR MEDIO DEL ID
data_complete = pd.merge(data, data_comp, how='inner', on='id')
data_complete.head(5)
```

SE COMIENZA CON EL TRATAMIENTO DE DATOS

```python
#ANTES SE REALIZA UNA COPIA DE SEGURIDAD
df = data_complete.copy()
df.shape
```

```python
#REVISIÓN DE NULOS
df.isnull().sum()
```

```python
#SE CAMBIA DE LAS COLUMNAS 'revenue' Y 'budget' LOS NULOS POR 0
df['revenue']= df['revenue'].replace([np.nan],['0'])
df['budget']= df['budget'].replace([np.nan],['0'])
```

```python
#SE ELIMINAN COLUMNAS QUE NO APORTAN INFORMACIÓN RELEVANTE AL DATAFRAME
df.drop(['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage'], axis=1, inplace=True)
df.shape
```

```python
#FUNCIÓN PARA NORMALIZAR COLUMNA CON DICCIONARIO Y PROCEDER A DESENCADENAR
def desanidar_dict(value):
    if pd.isnull(value):
        return ''
    else:
        collection = ast.literal_eval(value)
    return collection['name']
```

```python
df['belongs_to_collection']= df['belongs_to_collection'].apply(desanidar_dict)
df.head(5)
```

```python
#FUNCIÓN PARA NORMALIZAR COLUMNA CON LISTA DE DICCIONARIOS
def desanidar_list_dict(value):
    if pd.isnull(value):
        return ''
    else:
        list_dict = ast.literal_eval(value)
        list_names = [companie['name'] for companie in list_dict]
    return list_names
```

```python
#FUNCIÓN PARA NORMALIZAR COLUMNA CON LISTA DE DICCIONARIOS Y OBTENER SÓLO EL DIRECTOR DE
#LA COLUMNA crew
def desanidar_director(value):
    if pd.isnull(value):
        return ''
    else:
        L = []
        for i in ast.literal_eval(value):
         if i['job'] == 'Director':
            L.append(i['name'])
        return L
```

```python
#FUNCIÓN PARA NORMALIZAR COLUMNA CON LISTA DE DICCIONARIOS, EN LA QUE SÓLO TOMARÁ LOS 5 PRIMEROS
#NOMBRES DE LA COLUMNA CAST (ACTORES PRINCIPALES)
# No se implementó en esta ocasión, ya que no fue necesaria la columna para el modelo de entrenamiento
def actor_reduce_list(value):
    if pd.isnull(value):
        return None
    else:
        names = []
        j = 0
        collection = ast.literal_eval(value)
        for i in collection:
            if j <= 5:
                names.append(i['name'])
            j+=1
        return names
```

```python
df['genres']= df['genres'].apply(desanidar_list_dict)
df['production_companies']= df['production_companies'].apply(desanidar_list_dict)
df['production_countries']= df['production_countries'].apply(desanidar_list_dict)
df['spoken_languages']= df['spoken_languages'].apply(desanidar_list_dict)
df['cast']= df['cast'].apply(desanidar_list_dict)
df['crew']= df['crew'].apply(desanidar_director)
df.head(5)
```

```python
#SE VISUALIZA VALORES NULOS DE LAS COLUMNAS
df.info()
```

```python
#ELIMINACIÓN DE VALORES NULOS DE LA COLUMNA RELEASE_DATE
df=df.dropna(subset=['release_date'])
df.shape
```

```python
#CAMBIAR TIPO DE DATO DE COLUMNA 'release_date' A DATE
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
#SE AGREGA COLUMNA RELEASE_YEAR
df['release_year'] = df['release_date'].dt.year
```

```python
#SE VISUALIZA EL DATAFRAME PARA CONFIRMAR QUE LA COLUMNA SE ENCUENTRE CORRECTAMENTE CODEADA
df.head(5)
```

```python
#SE CAMBIA EL TIPO DE DATO DE LAS COLUMNAS Revenue y Budget PARA CREAR LA COLUMNA 'return'
df['revenue'] = df['revenue'].astype('float')
df['budget'] = df['budget'].astype('float')
```

```python
#SE GENERA LA COLUMNA RETURN
df['return'] = df['revenue']/df['budget']
```

```python
#CORRIGIENDO ERRORES SURGIDOS DURANTE LA DIVISIÓN
df['return']= df['return'].replace([np.nan],['0'])
df['return']= df['return'].replace([np.inf],['0'])
df['return'] = df['return'].astype('float')
```
